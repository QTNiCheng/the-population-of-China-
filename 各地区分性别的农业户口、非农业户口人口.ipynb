```python
import pandas as pd
import numpy as np
import pyecharts as pye
```

```python
data = pd.read_excel(r'data/1-2  各地区分性别的农业户口、非农业户口人口.xls')
data.head(5)
```

```python
attr = data.index[1:]
title = ['农业户口人口', '农业户口人口男', '农业户口人口女', '非农业户口人口', '非农业户口男', '非农业人口女']
v1 = [data['农业户口人口小计'][1:], data['农业户口人口男'][1:], data['农业户口人口女'][1:], data['非农业户口人口小计'][1:], data['非农业户口人口男'][1:], data['非农业户口人口女'][1:]]
v1 = [round(a/1000000, 2) for a in v1]
bar = pye.Bar()
bar.add(title[0], attr, v1[0], xaxis_rotate=70, xaxis_interval=0, is_stack=True)
bar.add(title[3], attr, v1[3], xaxis_rotate=70, xaxis_interval=0, is_stack=True)
bar_1 = pye.Bar()
bar_1.add(title[1], attr, v1[1], xaxis_rotate=70, xaxis_interval=0, is_stack=True)
bar_1.add(title[2], attr, v1[2], xaxis_rotate=70, xaxis_interval=0, is_stack=True)
bar_2 = pye.Bar()
bar_2.add(title[4], attr, v1[4], xaxis_rotate=70, xaxis_interval=0, is_stack=True)
bar_2.add(title[5], attr, v1[5], xaxis_rotate=70, xaxis_interval=0, is_stack=True)
page = pye.Page()
page.add(bar)
page.add(bar_1)
page.add(bar_2)
page
```

```python
attr = data.index
title = ['非农业户口人口占总人口比重', '非农业户口人口占总人口比重男', '非农业户口人口占总人口比重女']
v1 = [data['非农业户口人口占总人口比重合计'], data['非农业户口人口占总人口比重男'], data['非农业户口人口占总人口比重女']]
line = pye.Line()
for i,j in zip(title, v1):
    line.add(i, attr, j, xaxis_rotate=70, xaxis_interval=0, mark_point=['max', 'min'], yaxis_formatter="%")
line
```
