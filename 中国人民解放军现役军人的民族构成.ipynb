```python
import pandas as pd
import pyecharts as pye
```

```python
data = pd.read_excel(r'data/1-2  中国人民解放军现役军人的民族构成.xls')
data_1 = data.sort_values('人口数', ascending=False)
data_1
```

```python
attr = data['民    族'][2:]
v1 = data['人口数'][2:]
bar = pye.Bar('中国人民解放军现役军人的民族构成')
bar.add('各民族人口', attr, v1, xaxis_rotate=70, xaxis_interval=0, 
        is_datazoom_show=True, mark_point=['max', 'min'])
bar
```

# 现役解放军人民族分布
- 上图数据不包括汉族数据。
- 人数最多的是满族，最少的是德昂族。（内心os:好多民族没听过啊）


```python
attr = list(data_1['民    族'][1:10])
attr.append('其他民族')
v1 = list(data_1['人口数'][1:10])
v1.append(data_1['人口数'][:1] - sum(v1))
pie = pye.Pie('各族族现役军人占比')
pie.add('', attr, v1, radius=[40, 75], label_text_color=None,
        is_label_show=True, legend_orient='vertical', legend_pos='right')
pie
```
