```python
import pandas as pd
import pyecharts as pye
```

```python
data = pd.read_excel(r'data/1-9  各地区分性别的15岁及以上文盲人口.xls')
data.head()
```

```python
attr = data.index[1:]
v1 = round(data['文盲人口合计'][1:]/10000, 2)
bar = pye.Bar('各省文盲人口(万人)')
bar.add('', attr, v1, xaxis_rotate=40, xaxis_interval=0, mark_point=['max', 'min'])
bar
```

```python
pie = pye.Pie('全国文盲比例')
v1 = [data['文盲人口合计'][0], data['15岁及以上人口合计'][0] - data['文盲人口合计'][0]]
pie.add('', ['文盲', '非文盲'], v1, is_label_show=1, radius=[40, 75])
pie
```
