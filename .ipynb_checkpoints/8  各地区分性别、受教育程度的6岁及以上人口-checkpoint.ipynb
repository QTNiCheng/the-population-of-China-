```python
import pandas as pd
import pyecharts as pye
```

```python
data = pd.read_excel(r'data/1-8  各地区分性别、受教育程度的6岁及以上人口.xls')
data.rename(columns={'小计': '小计.0', '男': '男.0', '女': '女.0'}, inplace=True)
data.head() #6岁及以上人口 1未上过学	 2小    学  3初    中 4高    中 4大学专科  5大学本科  6研究生1
```

```python
attr = data.index
columns = data.columns #除去全国数据
label = ['未上学', '小学', '初中', '高中', '大专', '本科',  '研究生']
bar = pye.Bar('各地6周岁以上学历情况(百万人)', width=1800, height=600)
for i,j in zip(range(2, len(columns), 3), label):
#     print(j, columns[i])
#     print(len(attr), len(data[columns[i]][1:]))
    v1 = round(data[columns[i]][1:]/1000000, 2)
    bar.add(j, attr[1:], v1,
           xaxis_rotate=30, xaxis_interval=0, label_emphasis_textcolor='#000000',
           )
bar
```

```python
[not_school, primary_school, middle_school, high_scholl, college, university, graduate_student] = [round(data[columns[i]]/data[columns[0]]*100, 2) for i in range(3, 22, 3)]
data_list = [not_school, primary_school, middle_school, high_scholl, college, university, graduate_student]
line = pye.Line('学历比例')
for i,j in zip(data_list, label):
#     print(len(i), len(attr))
    line.add(j, attr, i,
            xaxis_rotate=30, xaxis_interval=0, label_emphasis_textcolor='#000000',)
# line.add('未上学', attr, not_school)
line
```

```python
pie = pye.Pie('各地区六周岁以上未上学比例图', width=1000, )
pie.add('', attr, not_school, radius=[40, 75], label_text_color=None,
        is_label_show=True, legend_orient='vertical', label_emphasis_textcolor='#000000',
        legend_pos='right')
pie
```

```python
v1 = [data['小计.{}'.format(i)][0] for i in range(1,8)]
# v1
pie = pye.Pie('全国学历比例图')
pie.add('', label, v1, radius=[40, 75], label_text_color=None,
        is_label_show=True, legend_orient='vertical',
        legend_pos='right')
pie
```

```python
v1 = [round(data['男.{}'.format(i)]/data['女.{}'.format(i)], 2)  for i in range(1, 8)]
v1
line = pye.Line('各地区学历男女比例', width=800,)
for i,j in zip(label, v1):
    line.add(i, attr, j, xaxis_rotate=30, xaxis_interval=0, label_emphasis_textcolor='#000000')
line
```
