```python
import pandas as pd
import numpy as np
import seaborn as sns
import pyecharts as pye
import decimal  
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
%matplotlib inline
```

```python
data = pd.read_excel(r'data/1-1  各地区户数、人口数和性别比.xls')
data.head()
```

```python
# data = data['合计'].sort_values()
x = list(data['地区'][1:])
y = (list(data['合计'][1:]/1000000))
y_ = [int(i) for i in y]
men = (list(data['男'][1:]/1000000))
men_ = [int(i) for i in men]
woman = (list(data['女'][1:]/1000000))
woman_ = [int(i) for i in woman]
# for i in y:
#     y_.append(int(i))
bar = pye.Bar("各省人总口及男女人口分布", width=1000, height=600)
bar.add('总人口（百万/人）', x, y_, 
        xaxis_rotate=90, label_color=['#000000'], 
        xaxis_interval=0,
        mark_point_symbolsize=60, is_label_show=True,
        label_text_color=('#0066ff')
       )
bar.add('女总人口（百万/人）', x, woman_, 
        is_stack=True, xaxis_rotate=90,
        xaxis_interval=0,
        mark_point_symbolsize=60, is_label_show=True
       )
bar.add('男总人口（百万/人）', x, men_, 
        is_stack=True, xaxis_rotate=90,
        xaxis_interval=0,
        mark_point_symbolsize=60, is_label_show=True
   )
bar
```

# 全国各省人口占比
- 最多人的是广东，意料中也是意料外，知道广东很多人，但是没想到居然是全国第一。以前一直以为河南才是全国第一的。
- 最少的是西藏，地理原因意料之中。

```python
attr = ["男", "女"]
v1 = [data['男'][0], data['女'][0]]
pie = pye.Pie("男女分布饼状图")
pie.add("", attr, v1, is_label_show=True)
pie
```

```python
data_t = data.T

attr_ = data_t[0:1].values
value_ = data_t[1:2].values
attr = []
for i in attr_:
    for i_ in i:
        attr.append(i_.replace(' ', ''))
value = []
for a in value_:
    for b in a:
        value.append(int(b/10000))
```

```python
map = pye.Map("全国人口分布热力图（/万人）", width=1000, height=800)
map.add("", attr, value, maptype='china', is_visualmap=True,
        visual_text_color='#000', visual_range =[0,11000],
       is_label_show=True)
map
```

# 全国各省人口地区分布图
- 图中可以很直观的看出全国各地人口分布的情况


```python
attr = list(data['地区'])
tem = (data['性别比（女100）']/100)
line = pye.Line('各地性别比')
v1 = []
for i in tem:
    v1.append(float( "%.2f" % i))
line.add('性别比(男/女)', attr, v1, xaxis_interval=0, 
         xaxis_rotate=90, yaxis_force_interval=0.05,
         yaxis_min=1, yaxis_max=1.2, mark_point=["max", "min"]
        )
line
```

```python
line = pye.Line('各地户口人数')
attr = list(data['地区'])
registered = data['（人/户）']
line.add('户口数（人/户）', attr, registered, xaxis_interval=0, 
         xaxis_rotate=90, yaxis_force_interval=0.5,mark_line=["average"],
          yaxis_min=1, yaxis_max=4.5, mark_point=["max", "min"]
        )
line
```
