```python
import pandas as pd
import numpy as np
import pyecharts as pye
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
%matplotlib inline
```

```python
data = pd.read_excel(r'data/各地别比（城、乡、镇）.xlsx')
data
```

# 预览数据


```python
attr = data.index
v1 = data['城（男女性别比）']/100
v2 = data['镇（男女性别比）']/100
v3 = data['乡（男女性别比）']/100
line = pye.Line('折线图', width=1000, height=500)
line.add('城（男女性别比）', attr, v1, yaxis_min=0.9, yaxis_max=1.2, xaxis_rotate=90, xaxis_interval=0, mark_point=["max", "min"])
line.add('镇（男女性别比）', attr, v2, yaxis_min=0.9, yaxis_max=1.2, xaxis_rotate=90, xaxis_interval=0, mark_point=["max", "min"])
line.add('乡（男女性别比）', attr, v3, yaxis_min=0.9, yaxis_max=1.2, xaxis_rotate=90, xaxis_interval=0, mark_point=["max", "min"])
line
```

# 城镇乡男女人口比例
- 图中可以看出城市户口男女比例最高的是我国大首都北京，比例最低的河北
- 而乡镇户口男女比例最高的依次是西藏和海南，最低的依次是重庆和江苏
- 男女比例低于1的有河北的城市户口和江苏的乡下户口（单身的汉子们可以考虑这两个地点的人，妹子多过汉子哦）

```python
data_1 = pd.read_excel(r'data/1-2  各地区分性别、户口登记状况的人口.xls')
data_1.head()
```

# 各地区分性别、户口登记状况的人口数据预览 ↑

```python
list_1 = round(data_1['居住本地小计']/data_1['合计'], 2)
plt.figure(figsize=(20, 10),)
plt.title('test')
for i,j,z in zip(data_1.index, list_1, range(1,33)):
    plt.subplot(4, 8, z)
    plt.pie([j, 1-j], autopct='%1.2f%%')
    plt.title(i)
    if z == 8:
        plt.legend(['居住本地', '在外'],bbox_to_anchor=(1.05, 1), loc='best', borderaxespad=0., fontsize=15)
```

# 各地区人口居住本地比例
- 从全国范围来看，有21%的人背井离乡，在外地打拼养家，中国依然有不少的留守儿童。
- 从图中所得，居住本地最少的居然是北京和上海，这两个国际化大都市本地人口比例都小于50%，难怪都说老北京老上海本地人不多。可能是因为发展起来了，人们有钱了，出国或者选择环境好的城市生活了吧。
- 西藏和贵州外出的人口比例最低，2个省都是我国经济相对不发达的省，为什么人们不喜欢外出呢？不太懂。

```python
pie = pye.Pie('人口现状分布', title_pos='center')
attr = ['居住本地', '离家半年以上', '国外工作或学习']
v1 = [data_1['居住本地小计'][:1], data_1['离开半年小计'][:1], data_1['国外工作或学习小计'][:1]]
pie.add('', attr, v1, radius=[40, 75], label_text_color=None,
        is_label_show=True, legend_orient='vertical',
        legend_pos='left')
pie
```

# 全国人口现状分布
- 全国有80.07%的人居住在本地，离家半年以上的有19.78%， 不在国内的（出国工作或读书）有0.15%。
- 出国的只有0.15%，也就是说我国差不多有200W人在国外，个人感觉比例有点低啊。
- 0.15%的比例，出国读书或工作是比较难的

```python
attr = data_1.index
v1 = [round(data_1["居住本地小计"]/data_1['合计']*100, 2), round(data_1["离开半年小计"]/data_1['合计'], 2)*100, round(data_1["国外工作或学习小计"]/data_1['合计']*100, 2)]
title = ['居住本地', '离开半年以上', '国外工作或学习']
line = pye.Line('居住比例图')
for i,j in zip(title, v1):
    line.add(i, attr,  j, xaxis_interval=0, xaxis_rotate=70, yaxis_formatter='%')
line
```

# 各地居住比例图
- 居住本地和来开半年以上没什么好看的，上图饼图中已经说明。比较有趣的是在国外的比例，吉林和福建是最高的这两省的人比较流行出国？亦或是其他什么原因呢？

```python
bar_2 = pye.Bar('居住情况柱状图')
attr = data_1.index[1:]
v1 = round(data_1['居住本地小计'][1:]/1000000, 2)
v2 = round(data_1['离开半年小计'][1:]/1000000, 2)
v3 = round(data_1['国外工作或学习小计'][1:]/10000, 2)
bar_2.add('居住本地(百万人)', attr, v1, xaxis_interval=0, xaxis_rotate=90, mark_point=['max', 'min'])
bar_2.add('离开半年(百万人)', attr, v2, xaxis_interval=0, xaxis_rotate=90, mark_point=['max', 'min'])
bar_2.add('国外工作或学习(万人)', attr, v3, xaxis_interval=0, xaxis_rotate=90, mark_point=['max', 'min'])
bar_2
```

# 居住情况柱状图
- 图中可以直观的看出各情况人口数值， 河南居住在本地的人口最多。最少的是西藏。（数值与本省人口有关）
- 离开本地半年以上人口最多的居然是广东，可能是以离开本市或本县计算的吧。人都到珠三角去了。最少的依然是西藏，还是人口总数的问题啊。
- 福建在国外的人口是全国最多的，看来福建人比较厉害啊。最少的依然是西藏。

```python
attr = data_1.index
v1 = round(data_1['居住本地男']/data_1['居住本地女'], 2)
v2 = round(data_1['离开半男']/data_1['离开半女'], 2)
v3 = round(data_1['男.1']/data_1['女.1'], 2)
line = pye.Line('居住情况男女比例折线图')
lables_list = ['居住本地男女比', '离开半年以上男女比', '出国男女比']
v_list = [v1, v2, v3]
for i,j in zip(lables_list, v_list):
    line.add(i, attr, j, xaxis_interval=0, xaxis_rotate=90,  mark_point=["max", "min"],  mark_line=["average"])
line
```

# 居住情况各地男女比例折线图
- 图中很容易看出居住在本地男女比例变化不大，趋近与直线。
- 但是离家在外的比例有点涨幅，比例最高的是天津， 最低的是广西。
- 出国的男女比例就涨幅就有点大了，比例最高的是安徽，最低的是海南。
