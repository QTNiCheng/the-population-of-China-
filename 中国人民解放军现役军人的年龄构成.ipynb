```python
import pandas as pd
import numpy as np
import pyecharts as pye
%matplotlib inline
```

```python
data = pd.read_excel(r'data/1-1  中国人民解放军现役军人的年龄构成.xls')
data
```

```python
attr = ["男", "女"]
v1 = [data['男'][0], data['女'][0]] #数据上面有且数据只有2个，直接把数据粘贴过来就比较方便了
pie = pye.Pie("我国男女兵比例", width=400, height=400)
pie.add("", attr, v1, is_label_show=True)
pie
```

# 男女兵比例图
- 我国男兵还是占绝对的比例的，毕竟自古以来保家卫国都是汉子居多。但是女兵有4.26%是出乎我意料的，巾帼不让须眉啊。

```python
attr = data['年  龄'][1:]
v1 = data['合  计'][1:]
v2 = data['男'][1:]
v3 = data['女'][1:]
line = pye.Line("折线图")
line.add("各阶段年龄人数", attr, v1,
         mark_point=["max", "min"], mark_line=["average"], xaxis_interval=0, xaxis_rotate=90)
line.add("男", attr, v2,
         mark_point=["max", "min"], mark_line=["average"], xaxis_interval=0, xaxis_rotate=90)
line.add("女", attr, v3,
         mark_point=["max", "min"], mark_line=["average"], xaxis_interval=0, xaxis_rotate=90)
line
```

# 各阶段年龄人数折线图
- 总的来说，是18到25岁之间人数最多，这是因为我国义务兵的原因吧。40岁以后数量就很少了，但是他们都算是大人物了吧，有经验的老兵了。
- 男兵的分布和全体分布几乎一致。
- 女性的分布也差不多。

```python
attr =  data['年  龄'][1:]
v1 = data['合  计'][1:]
v2 = data['男'][1:]
v3 = data['女'][1:]
bar = pye.Bar('柱状图', width='1000', height=600)
bar.add('总数', attr, v1, xaxis_rotate=90)
bar.add('男', attr, v2, xaxis_rotate=90, is_stack=True)
bar.add('女', attr, v3, xaxis_rotate=90, is_stack=True)
bar
```

# 年龄分布柱状图，可以直观的看出数值
- 比例在上面的折线图都能看出来，柱状图是可以看出个阶段的数值
